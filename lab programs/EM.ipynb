```python
from sklearn.mixture import GaussianMixture
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
import matplotlib.pyplot as plt
model = GaussianMixture(n_components=2,covariance_type="spherical",random_state=4)

X,y = make_blobs(n_samples=1000,n_features=2,centers=2,random_state=4,cluster_std=1.2)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=4)

model.fit(X_train)

print(accuracy_score(y_test,model.predict(X_test)))
print(confusion_matrix(y_test,model.predict(X_test)))

plt.scatter(X_test[:,0],X_test[:,1],s=10,c=y_test,cmap="viridis")
plt.title('Synthetic Dataset')
plt.show()


plt.scatter(X_test[:, 0], X_test[:, 1], c=model.predict(X_test), s=10, cmap='viridis')
plt.title('EM Clustering Results')
plt.show()
```
