```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler,RobustScaler
import numpy as np
import pandas as pd
n=int(input())
input()
features1=[]
features2=[]
target1=[]
target2=[]
order=[]
data_column1_missing=[]
data_column2_missing=[]
for i in range(n):
    record=input().split('\t')
    record[0]=int(record[0])
    if "Missing" in record[2]:
        record[3]=float(record[3])
        data_column1_missing.append(record[:2]+record[3:])
        order.append(0)
    elif "Missing" in record[3]:
        record[2]=float(record[2])
        data_column2_missing.append(record[:3])
        order.append(1)
    else:
        features1.append(record[:3])
        features2.append(record[:2]+record[3:])
        record[2]=float(record[2])
        record[3]=float(record[3])
        target1.append(record[2])
        target2.append(record[3])

encoder = LabelEncoder()
features1=pd.DataFrame(data=features1,columns=["feature1","feature2","feature3"])
features1["feature2"]=encoder.fit_transform(features1["feature2"])

features2=pd.DataFrame(data=features2,columns=["feature1","feature2","feature3"])
features2["feature2"]=encoder.fit_transform(features2["feature2"])

features1=np.array(features1)
features2=np.array(features2)
scaler=RobustScaler()
""" features1=scaler.fit_transform(features1)
features2=scaler.fit_transform(features2) """
target1=pd.DataFrame(data=target1)
target2=pd.DataFrame(data=target2)

data_column1_missing=pd.DataFrame(data=data_column1_missing,columns=["feature1","feature2","feature3"])
data_column2_missing=pd.DataFrame(data=data_column2_missing,columns=["feature1","feature2","feature3"])

data_column1_missing["feature2"]=encoder.fit_transform(data_column1_missing["feature2"])
data_column2_missing["feature2"]=encoder.fit_transform(data_column2_missing["feature2"])

model1=KNeighborsRegressor(n_neighbors=4)
model1.fit(features1,target1)

model2=KNeighborsRegressor(n_neighbors=4)
model2.fit(features2,target2)

p1=0
p2=0

""" data_column1_missing=scaler.transform(np.array(data_column1_missing))
data_column2_missing=scaler.transform(np.array(data_column2_missing)) """

data_column1_missing=np.array(data_column1_missing)
data_column2_missing=np.array(data_column2_missing)
preds1=model1.predict(data_column1_missing)
preds2=model2.predict(data_column2_missing)
""" print(preds1)
print(order) """
#print(features[:,1])
print(model1.score(features1,target1))
print(model2.score(features2,target2))
pred=""
for i in range(len(data_column1_missing)+len(data_column2_missing)):
    if order[i]==0:
        pred=preds1[p1][0]
        p1+=1
    else:
        pred=preds2[p2][0]
        p2+=1
    print(str(round(pred,1)))
```
